```python
# download stopwords corpus, you need to run it once
import nltk
nltk.download("stopwords")

from nltk.corpus import stopwords
import pymorphy2
from string import punctuation
```

```python
print(stopwords.fileids()) # К сожалению в nltk нет поддержки русского языка но мы 
# можем создать свой кастомный корпус слов для украинского или можно найти другие источники
```

```python
#Create lemmatizer and stopwords list
morph = pymorphy2.MorphAnalyzer()
russian_stopwords = stopwords.words("russian")
```

```python
from collections import defaultdict
```

```python
def generate_ngrams(text, n_gram=1):
    text = str(text)
    tokens1 = [token for token in text.lower().split(' ') if token != '' if token not in russian_stopwords]
    tokens2 =[]
    for token in tokens1:
        token = morph.parse(token)[0].normal_form
        token = token.translate(str.maketrans('', '', punctuation))
        tokens2.append(token)
    ngrams = zip(*[tokens2[i:] for i in range(n_gram)])
    return [' '.join(ngram) for ngram in ngrams]
```

```python
def generate_patterns(n_gram=2):
  alem_bigrams = defaultdict(int)

  for tweet in df_marafon['text']:
      for word in generate_ngrams(tweet,n_gram):
          alem_bigrams[word] += 1

  alem_bigrams = pd.DataFrame(sorted(alem_bigrams.items(), key=lambda x: x[1])[::-1])
  return generate_patterns(n_gram=2)
```
